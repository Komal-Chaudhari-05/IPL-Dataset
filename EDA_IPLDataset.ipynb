```python
# Extracting the records where a team won batting first
batting_first = ipl[ipl['win_by_runs'] != 0]
batting_first
```

```python
batting_first.head()
```

```python
# making a histogram
plt.figure(figsize=(5,5))  # size of histogram
plt.hist(batting_first['win_by_runs'])  # use win_by_runs column from batting_first datafream
plt.title('Distribution of Runs')  # title of histogram plot
plt.xlabel('Runs')
plt.show()
```

```python
# Finding out the number of wins each team after batting first
batting_first['winner'].value_counts()  
```

```python
# Making a Bar Plot for top 3 teams with most wins after batting first
plt.figure(figsize=(3,3))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()), list(batting_first['winner'].value_counts()[0:3]), color=['green','blue','red'])
```

```python
# Making pie chart
plt.figure(figsize=(5,5))
plt.pie(list(batting_first['winner'].value_counts()), labels=list(batting_first['winner'].value_counts().keys()), autopct='%0.1f%%')
 #In pie plot if first parameter is numarical then the other one is categorical     Autopct - it is used for displayed percentage distribution 
```
